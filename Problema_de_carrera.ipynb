```python
class Calculadora:
    def __init__(self, capital: float, tasa: float, tiempo: float):
        self.capital = capital
        self.tasa = tasa / 100
        self.tiempo = tiempo

    def interes_simple(self):
        interes = self.capital * self.tasa * self.tiempo
        total = self.capital + interes
        return interes, total

    def interes_compuesto(self):
        total = self.capital * (1 + self.tasa) ** self.tiempo
        interes = total - self.capital
        return interes, total


def pedir_dato(mensaje):
    while True:
        try:
            valor = float(input(mensaje))
            if valor < 0:
                print("⚠️ Ingresa un valor positivo.")
                continue
            return valor
        except ValueError:
            print("❌ Entrada inválida. Intenta de nuevo.")


def mostrar_resultados(tipo, interes, total):
    print(f"\n--- {tipo} ---")
    print(f"Interés generado: ${interes:,.2f}")
    print(f"Total: ${total:,.2f}")


def main():
    print("📊 CALCULADORA DE INTERÉS SIMPLE Y COMPUESTO\n")

    capital = pedir_dato("Capital inicial: ")
    tasa = pedir_dato("Tasa de interés anual (%): ")
    tiempo = pedir_dato("Tiempo (años): ")

    calc = Calculadora(capital, tasa, tiempo)

    interes_s, total_s = calc.interes_simple()
    interes_c, total_c = calc.interes_compuesto()

    mostrar_resultados("INTERÉS SIMPLE", interes_s, total_s)
    mostrar_resultados("INTERÉS COMPUESTO", interes_c, total_c)


if __name__ == "__main__":
    main()
```
