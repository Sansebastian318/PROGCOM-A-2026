````
suma=0
c=0
notas=[]
n=int(input("Ingrese la cantidad de estudiantes"))
for i in range(n):
  nota=float(input(f"Ingrese la nota del estudiante {i}:"))
  notas.append(nota)
  suma+=nota
prom=suma/n
for i in range(n):
   if notas [i]>prom:
      c=c+1
      print(f"{c}estudiantes sacaron mayor nota que el promedio")
````

````
# Lista de números de una lista son múltipos de número
numeros = [9, 15, 20, 21, 30, 33, 40, 48, 52, 54, 60,99, 100]
n = int(input("Ingrese el número n: "))
contador = 0
for num in numeros:
    if num % n == 0:
        contador += 1
print("Cantidad de múltiplos de", n, ":", contador)
````
