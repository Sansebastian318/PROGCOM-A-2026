##Tabla de multiplicar hasta el 10

```
n= int(input("Ingrese número: "))
for i in range (1, 11):
  print(n, "x", i, "=", n*i)
```

##Suma de número naturales hasta n

```
n = int(input("¿Hasta qué número desea sumar? (n): "))

suma = 0
i = 1

while i <= n:
    suma += i
    i += 1

print("La suma de los números naturales hasta", n, "es:", suma)
```

##Factorial

```
n= int(input("Ingrese el número que desee "))
if n < 0:
  print("No existe factorial negativo")
else:
    factorial = 1
    i=1
    while i <= n:
      factorial *= i
      i +=1
      print("Factorial: ", factorial)
```

##Serie Fibonacci

```
n= int(input("Cantidad de términos: "))
a,b = 0,1
for i in range(n):
  print(a)
  a,b =b, a+b
```

##Adivinar número, máximo 10 intentos

```
import random
numero = random.randint(1,10)
intentos = 0
while intentos < 10:
  guess = int(input("Adivina (1-10): "))
  intentos += 1
  if guess == numero:
    print("¡Correcto!")
    break
else:
 print("Perdiste. Era:", numero)
```

##Suma de dígitos

```
n = int(input("Número: "))
suma = 0
while n > 0:
  suma += n %10
  n //=10
  print("Suma de digitos: ", suma)
```

##Número primo

```
n = int(input("Número: "))
es_primo = True
if n < 2:
    es_primo = False
for i in range(2, n):
 if n % i == 0:
  es_primo = False
 break
print("Es primo" if es_primo else "No es primo")
```
