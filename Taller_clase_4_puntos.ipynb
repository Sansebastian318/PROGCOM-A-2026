```
n = int(input("Cantidad de trabajadores: "))
nombres= []
salarios= []
c=0
for i in range(n):
    nombre = input("Nombre del trabajador: ")
    salario = float(input("Salario: "))
    nombres.append(nombre)
    salarios.append(salario)
promedio = sum(salarios) / n
print("Salario promedio:", promedio)
print("Trabajadores que ganan más que el promedio:")
contador = 0
for i in range(n):
    if salarios[i] > promedio:
        print(nombres[i])
        contador + 1
```

```
# Lista de números de una lista son múltipos de número
numeros = [9, 15, 20, 21, 30, 33, 40, 48, 52, 54, 60,99, 100]
n = int(input("Ingrese el número n: "))
contador = 0
for num in numeros:
    if num % n == 0:
        contador += 1
print("Cantidad de múltiplos de", n, ":", contador)
```
