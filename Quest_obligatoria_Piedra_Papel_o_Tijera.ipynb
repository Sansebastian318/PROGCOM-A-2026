```
import random
print("Juego: Piedra, Papel, Tijera, Spoke o Lagarto envenenado")
print("☢️Debes tener en cuenta estas reglas: Spock rompe tijera, tijera decapita lagarto, tijera corta papel, papel desautoriza Spock, piedra aplasta tijera, papel envuelve piedra, piedra aplasta lagarto, lagarto devora papel, lagarto envenena Spock, Spock vaporiza piedra.")
print("1 = Piedra 🪨")
print("2 = Papel 📄")
print("3 = Tijera ✂️")
print("4 = Spock 🖖")
print("5 = Lagarto envenenado 🦎")
while True:
    while True:
        try:
            yo = int(input("Elige 1, 2 ,3, 4 o 5: "))
            if yo not in [1, 2, 3, 4, 5]:
                print("❌ Solo puedes elegir 1, 2, 3, 4 o 5.\n")
                continue
            break
        except ValueError:
            print("❌ Debes ingresar un número válido.\n")
    m = random.randint(1, 5)
    opciones = {
        1: "Piedra 🪨",
        2: "Papel 📄",
        3: "Tijera ✂️",
        4: "Spock 🖖",
        5: "Lagarto envenenado 🦎"
    }
    print(f"\nTú elegiste: {opciones[yo]}")
    print(f"Máquina eligió: {opciones[m]}")
    if yo == m:
        print("Empate 🤝")
    elif (yo == 1 and m == 3) or (yo == 2 and m == 1) or (yo == 3 and m == 2) or (yo == 4 and m == 1) or (yo == 5 and m == 4) or (yo == 1 and m == 5) or (yo == 2 and m == 4) or (yo == 3 and m == 5) or (yo == 4 and m == 3) or (yo == 5 and m == 2):
        print("¡GANASTE! 🎉")
    else:
        print("Perdiste 😢")
    continuar = input("¿Deseas continuar si/no? ").lower()
    if continuar == "no":
        print("Gracias por jugar.")
        break
```
