2.Imprimir los votos obtenidos por 5 candidatos distintos, Suponiendo que vontan n usuarios. Seleccione los dos primeros y vuelva a ejecutar la votacion con solo esos dos (segunda vuelta).

```
suma = 0
c = 0
notas = []
n = 10
for i in range(n):
    nota = float(input(f"Ingrese la nota del estudiante {i+1}: "))
    notas.append(nota)
    suma += nota
prom = suma / n
print(f"\nPromedio del curso: {prom:.2f}")
print(f"\nEstudiantes con nota mayor al promedio:")
for i in range(n):
    if notas[i] > prom:
        c += 1
        print(f" - Estudiante {i+1}: {notas[i]}")
print(f"\nTotal: {c} estudiantes sacaron mas que el promedio")
nota_max = max(notas)
indice_max = notas.index(nota_max)
print(f"\nEl estudiante con mayor nota fue el estudiante {indice_max+1} con {nota_max}")
```
