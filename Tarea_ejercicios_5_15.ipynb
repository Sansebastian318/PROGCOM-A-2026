# Ejercicio con ecuaciones cuadráticas

```
import math
a = float(input("Ingrese valor de a: "))
b = float(input("Ingrese valor de b: "))
c = float(input("Ingrese valor de c: "))
d = b**2 -4*a*c
print(d)
if d > 0:
  x1 = (-b + math.sqrt(d))/(2*a)
  x2 = (-b - math.sqrt(d))/(2*a)
  print("Hay dos soluciones: ", x1, x2)
elif d == 0:
    x = -b/(2*a)
    print("Hay una solución", x)
else:
    print("No tiene soluciones reales")
```
