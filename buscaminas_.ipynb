```python
import tkinter as tk
import random

FILAS = 8
COLUMNAS = 8
MINAS = 10

tablero = []
botones = []
banderas = set()
puntaje = 0
juego_terminado = False

COLORES = {
    1: "blue",
    2: "green",
    3: "red",
    4: "darkblue",
    5: "brown",
    6: "turquoise",
    7: "black",
    8: "gray"
}

def crear_tablero():
    global tablero
    tablero = [[0 for _ in range(COLUMNAS)] for _ in range(FILAS)]

    minas = 0
    while minas < MINAS:
        f = random.randint(0, FILAS - 1)
        c = random.randint(0, COLUMNAS - 1)
        if tablero[f][c] != -1:
            tablero[f][c] = -1
            minas += 1

    for f in range(FILAS):
        for c in range(COLUMNAS):
            if tablero[f][c] == -1:
                continue
            count = 0
            for i in range(-1, 2):
                for j in range(-1, 2):
                    nf, nc = f + i, c + j
                    if 0 <= nf < FILAS and 0 <= nc < COLUMNAS:
                        if tablero[nf][nc] == -1:
                            count += 1
            tablero[f][c] = count

def click_izq(f, c):
    global puntaje, juego_terminado

    if juego_terminado or (f, c) in banderas:
        return

    if tablero[f][c] == -1:
        botones[f][c].config(text="💣", bg="red")
        fin_juego(False)
        return

    valor = tablero[f][c]

    if valor == 0:
        botones[f][c].config(text="", bg="lightgray")
    else:
        botones[f][c].config(text=str(valor), fg=COLORES.get(valor, "black"), bg="lightgray")

    botones[f][c].config(relief="sunken", state="disabled")

    puntaje += 1
    actualizar_puntaje()

def click_der(event, f, c):
    if juego_terminado:
        return

    if botones[f][c]["state"] == "disabled":
        return

    if (f, c) in banderas:
        botones[f][c].config(text="")
        banderas.remove((f, c))
    else:
        botones[f][c].config(text="F")
        banderas.add((f, c))

def fin_juego(gano):
    global juego_terminado
    juego_terminado = True

    for f in range(FILAS):
        for c in range(COLUMNAS):
            if tablero[f][c] == -1:
                botones[f][c].config(text="💣")

    if gano:
        estado_label.config(text="Ganaste")
    else:
        estado_label.config(text="Perdiste")

def actualizar_puntaje():
    puntaje_label.config(text="Puntaje: " + str(puntaje))

ventana = tk.Tk()
ventana.title("Buscaminas")

frame = tk.Frame(ventana)
frame.pack()

for f in range(FILAS):
    fila = []
    for c in range(COLUMNAS):
        b = tk.Button(frame, width=4, height=2,
                      command=lambda f=f, c=c: click_izq(f, c))
        b.grid(row=f, column=c)
        b.bind("<Button-3>", lambda e, f=f, c=c: click_der(e, f, c))
        fila.append(b)
    botones.append(fila)

puntaje_label = tk.Label(ventana, text="Puntaje: 0")
puntaje_label.pack()

estado_label = tk.Label(ventana, text="En juego")
estado_label.pack()

crear_tablero()
ventana.mainloop()
```
