# ESTRUCTURAS DE REPETICIÓN


##1. PARA - FOR
* Se emplea cuando se sabe cuántas veces se repetirá la acción.
* **OJO** El conteo siempre inicia desde **0**.
Sintaxis general

```
for (variable) in (algo):
La acción que se va a repetir

#Forma2
for variable in range(cantidad de veces):
La acción que se va a repetir

#Forma 3
for variable in range(VI,VF+1,P):
La acción que se va a repetir
```




Donde:
* VI: Valor inicial (donde inicio).
* VF: Valor final (dónde finalizó)
* P: Paso. Cada cuánto cuento

sintaxis para java

```
for (int variable=VI; variable= VF;variable=variable+paso){la acción que se va a repetir}
```
