````
import random
print("Juego: Piedra, Papel o Tijera")
print("1 = Piedra")
print("2 = Papel")
print("3 = Tijera")
while True:
    try:
        yo = int(input("Elige 1, 2 o 3: "))
        if yo not in [1, 2, 3]:
            print("❌ Solo puedes elegir 1, 2 o 3.\n")
            continue
        break  # Si todo está bien, sale del ciclo
    except ValueError:
        print("❌ Debes ingresar un número válido (1, 2 o 3).\n")
m = random.randint(1, 3)

opciones = {1: "Piedra", 2: "Papel", 3: "Tijera"}
print(f"\nTú elegiste: {opciones[yo]}")
print(f"Máquina eligió: {opciones[m]}")
if yo == m:
    print("Empate 🤝")
elif (yo == 1 and m == 3) or (yo == 2 and m == 1) or (yo == 3 and m == 2):
    print("¡GANASTE! 🎉")
else:
    print("Perdiste 😢")
````
